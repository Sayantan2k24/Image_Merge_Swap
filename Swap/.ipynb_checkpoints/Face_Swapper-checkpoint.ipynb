```python
import matplotlib.pyplot as plt
import cv2
import time


def load_img(img):
    image = cv2.imread(img)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image


def render(img_obj):
    plt.imshow(img_obj)
    plt.show()


def swap_faces(img_obj1, img_obj2):
    # Define hardcoded regions for swapping (replace these values with your own regions)
    # Region from the first image to be swapped with the second image
    x1 = img_obj1[:100, 50:200]  # Replace these coordinates to capture the face region
    # Region from the second image to be swapped with the first image
    y1 = img_obj2[:300, 190:500]  # Replace these coordinates to capture the face region

    # Resize the regions to match each other
    face_region2_resized = cv2.resize(x1, (y1.shape[1], y1.shape[0]))

    # Swap the face regions
    img_obj1[:100, 50:200] = face_region2_resized
    img_obj2[:300, 190:500] = y1

    return img_obj1, img_obj2


if __name__ == "__main__":
    print("Loading the first image...")
    img_obj1 = load_img("image1.png")  # dore
    render(img_obj1)
    time.sleep(2)

    print("Loading the second image...")
    img_obj2 = load_img("image2.png")  # man
    render(img_obj2)
    time.sleep(2)

    print("Swapping the faces...")
    s1, s2 = swap_faces(img_obj1.copy(), img_obj2.copy())
    time.sleep(2)

    print("Displaying the swapped faces...")
    print("Swapped face 1:")
    render(s1)
    time.sleep(2)

    print("Swapped face 2:")
    render(s2)
    time.sleep(2)
    print("Done!")
```
